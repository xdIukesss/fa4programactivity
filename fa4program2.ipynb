```python
correct_password = "7p3@rl"
attempts = 3

while attempts > 0:
    password = input("Enter password: ")

    if password == correct_password:
        print("Access Granted")
        break
    else:
        attempts -= 1
        print("Incorrect password. Attempts left:", attempts)

if attempts == 0:
    print("Access Denied")
```
